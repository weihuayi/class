## § 5.1 线性微分方程组解的存在唯一性定理
## Existence & Uniqueness Theorems of Linear ODEs

### 本节要求 /Requirements/
* 掌握高阶线性微分方程与线性微分方程组的关系.
* 理解线性微分方程组解的存在唯一性定理.
* 熟练掌握解的逐次逼近序列的构造方法.

### 5.1.1 记号与定义 /Symbol and Definition/

一阶微分方程组

$$
\begin{cases}
x_1'=f_1(t,x_1,x_2,\dots,x_n)\\
x_2'=f_2(t,x_1,x_2,\dots,x_n)\\
\dots\dots\dots\dots\\
x_n'=f_n(t,x_1,x_2,\dots,x_n)\\
\end{cases}
$$

初值条件 $x_1(t_0)=\eta_1$ , $x_2(t_0)=\eta_2$ , $\dots$ , $x_n(t_0)=\eta_n$

一阶线性微分方程组

$$
\begin{cases}
x_1'=a_{11}(t)x_1+a_{12}(t)x_2+\dots+a_{1n}(t)x_n+f_1(t)\\
x_2'=a_{21}(t)x_1+a_{22}(t)x_2+\dots+a_{2n}(t)x_n+f_2(t)\\
\dots\dots\dots\dots\\
x_n'=a_{n1}(t)x_1+a_{n2}(t)x_2+\dots+a_{nn}(t)x_n+f_n(t)\\
\end{cases}
$$

其中 $a_{ij}$, $f_i(t)$ $\quad$ $i,j=1,2,\dots,n$ 在 $[a,b]$ 上连续.

$$
\mathbf A(t)=
\left[
\begin{matrix}
a_{11}(t)&a_{12}(t)&\dots&a_{1n}(t)\\
a_{21}(t)&a_{22}(t)&\dots&a_{2n}(t)\\
\dots&\dots&\dots&\dots\\
a_{n1}(t)&a_{n2}(t)&\dots&a_{nn}(t)\\
\end{matrix}
\right]
$$

$$
\mathbf f(t)=
\left[
\begin{matrix}
f_1(t)\\
f_2(t)\\
\vdots\\
f_n(t)\\
\end{matrix}
\right]
\quad
\mathbf x=
\left[
\begin{matrix}
x_1\\
x_2\\
\vdots\\
x_n\\
\end{matrix}
\right]
\quad
\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=
\left[
\begin{matrix}
x_1'\\
x_2'\\
\vdots\\
x_n'\\
\end{matrix}
\right]
$$

$$
\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=\mathbf A(t)\mathbf x+\mathbf f(t)
$$

* **矩阵和向量函数性质与运算** 

在区间 $a\leq t\leq b$ 可定义矩阵与向量函数
$$
\mathbf B(t)=
\left[
\begin{matrix}
b_{11}(t)&b_{12}(t)&\dots&b_{1n}(t)\\
b_{21}(t)&b_{22}(t)&\dots&b_{2n}(t)\\
\dots&\dots&\dots&\dots\\
b_{n1}(t)&b_{n2}(t)&\dots&b_{nn}(t)\\
\end{matrix}
\right]
\quad
\mathbf u(t)=
\left[
\begin{matrix}
u_1(t)\\
u_2(t)\\
\vdots\\
u_n(t)\\
\end{matrix}
\right]
$$

**连续:** $b_{ij}(t)$ , $u_i(t)$ 在区间 $a\leq t\leq b$ 连续.

**可微:** $b_{ij}(t)$ , $u_i(t)$ 在区间 $a\leq t\leq b$ 可微.

$\mathbf B(t)'=(b_{ij}'(t))_{n\times n}$ $\quad$ $u(t)'=(u_1'(t),u_2'(t),\dots,u_n'(t))^T$

**可积:** $b_{ij}(t)$ , $u_i(t)$ 在区间 $a\leq t\leq b$ 可积.

$\int \mathbf B(t){\mathrm d t}=\left(\int b_{ij}(t){\mathrm d t}\right)_{n \times n}$

$\int \mathbf u(t){\mathrm d t}=\left(\int u_1(t){\mathrm d t},\int u_2(t){\mathrm d t},\dots,\int u_n(t){\mathrm d t}\right)^T$

(1) $\left(\mathbf A(t)+\mathbf B(t)\right)'=\mathbf A'(t)+\mathbf B'(t)$ 

(2) $\left(\mathbf u(t)+\mathbf v(t)\right)'=\mathbf u'(t)+\mathbf v'(t)$

(3) $\left(\mathbf A(t)\mathbf B(t)\right)'=\mathbf A'(t)\mathbf  B(t)+\mathbf B'(t)\mathbf A(t)$

(4) $\left(\mathbf A(t)\mathbf u(t)\right)'= \mathbf A'(t) \mathbf u(t)+ \mathbf A(t)\mathbf u'(t)$

**定义 1 **设 $\mathbf A(t)$ 是区间 $a\leq t\leq b$ 上的连续 $n\times n$ 矩阵,　$\mathbf f(t)$ 是区间 $a\leq t\leq b$ 上的连续 $n$ 维向量, 方程组

$$
\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=\mathbf A(t)\mathbf x+ \mathbf f(t)
$$

称连续向量函数 $\mathbf u(t)$ 在某区间 $\alpha \leq t \leq \beta $ $\left([\alpha,\beta]\subset[a,b]\right)$ 是方程组的**解**, 则 $\mathbf u(t)$ 满足方程

$$
\mathbf u'(t)=\mathbf A(t)\mathbf u(t)+\mathbf f(t)
$$

**定义 2** 初值问题(Cauchy Problem)

$$
\begin{cases}
\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=\mathbf A(t)\mathbf x+\mathbf f(t)\\
\mathbf x(t_0)=\mathbf \eta\\
\end{cases}
$$

的解就是方程组 $\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=\mathbf A(t)\mathbf x+\mathbf f(t)$ 在包含 $t_0$ 的区间 $\alpha \leq t \leq \beta $ 上的解 $\mathbf u(t)$ , 使得 $\mathbf u(t_0)=\eta$.

### 5.1.2 n 阶线性微分方程与一阶线性微分方程组等价

**例 1**
$$
x''+p(t)x'+q(t)x=f(t)\\
x(t_0)=\eta_1, x'(t_0) = \eta_2
$$

**解:** 

令 $x_1=x$, $x_2=x'$, 则

$$ 
x_2'=-p(t)x'-q(t)x+f(t) 
$$

$$
\begin{cases}
x_1'=x_2\\
x_2'=-p(t)x_2-q(t)x_1+f(t)\\
\end{cases}
$$

$$
\mathbf x'=
\left[
\begin{matrix}
0&1\\
-q(t)&-p(t)\\
\end{matrix}
\right]\mathbf x+
\left[
\begin{matrix}
0\\
f(t)\\
\end{matrix}
\right]
$$

$$
x_1(t_0)=\eta_1,x_2(t_0)=\eta_2
$$

$$
\mathbf x'=
\left[
\begin{matrix}
0&1\\
-q(t)&-p(t)\\
\end{matrix}
\right]\mathbf x+
\left[
\begin{matrix}
0\\
f(t)\\
\end{matrix}
\right]
$$

下面考虑 $n$ 阶线性微分方程的初值问题

$$
x^{(n)}+a_1(t)x^{(n-1)}+\dotsi+a_{n-1}(t)x'+a_{n}(t)x=f(t)
$$

令

$$
x_1=x, x_2=x', x_3=x'',\dotsi,x_n=x^{(n-1)}
$$

则

$$
\begin{cases}
x_1'=x'=x_2\\
x_2'=x''=x_3\\
\dotsi\\
x_n'=x^{(n)}=-a_n(t)x_1-a_{n-1}(t)x_2-\dotsi-a_1(t)x_n+f(t)
\end{cases}
$$

写成矩阵形式
$$
\mathbf x'=
\begin{bmatrix}
  0 & 1 & 0 & \dots & 0 & 0\\
  0 & 0 & 1 & \dots & 0 & 0\\
\vdots & \vdots & \vdots & \ddots & \vdots & \vdots\\
  0 & 0 & 0 & \dots & 1 & 0\\
0 & 0 & 0 & \dots & 0 & 1\\
-a_n{(t)}&-a_{n-1}{(t)}&-a_{n-2}(t)&\dotsi&-a_2{(t)}&-a_1{(t)}\\
\end{bmatrix}\mathbf x+
\begin{bmatrix}
0\\
0\\
\dots\\
0\\
f(t)\\
\end{bmatrix}
$$


方程组加入初值条件 
$$
x(t_0)=\eta_1,x'(t_0)=\eta_2,\dotsi,x^{(n-1)}(t_0)=\eta_n
$$

$$
\mathbf x'=
\begin{bmatrix}
0&1&0 &\dots&0&0\\
0&0&1 &\dots&0&0\\
\vdots&\vdots&\vdots&\ddots&\vdots&\vdots\\
0&0&0&\dots&1&0\\
0&0&0&\dots&0&1\\
-a_n{(t)}&-a_{n-1}{(t)}&-a_{n-2}(t)&\dotsi&-a_2{(t)}&-a_1{(t)}\\
\end{bmatrix}
\mathbf x+
\left[
\begin{matrix}
0\\
0\\
\dots\\
0\\
f(t)\\
\end{matrix}
\right]
$$

$$
\mathbf x'(t_0)=
\left[
\begin{matrix}
\eta_1\\
\eta_2\\
\dots\\
\eta_n\\
\end{matrix}
\right]
$$

### 5.1.3 存在唯一性定理

初值问题(Cauchy Problem)

$$
\begin{cases}
\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=\mathbf A(t)\mathbf x+\mathbf f(t)\\
\mathbf x(t_0)=\mathbf \eta\\
\end{cases}
$$

**定理 1** 如果 $\mathbf A(t)$ 是 $n\times n$ 矩阵, $\mathbf f(t)$是 $n$ 维列向量,它们都在区间 $a\leq t\leq b$ 上连续, 则对于区间 $a\leq t\leq b$ 上的任何数 $t_0$ 及任一常数向量

$$
\mathbf x{(t_0)}=
\left[
\begin{matrix}
\eta_1\\
\eta_2\\
\vdots\\
\eta_n\\
\end{matrix}
\right]=\mathbf \eta
$$

方程组

$$
\begin{cases}
\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=\mathbf A(t)\mathbf x+\mathbf f(t)\\
\mathbf x(t_0)=\mathbf \eta\\
\end{cases}
$$

存在唯一解 $\mathbf \varphi(t)$ 定义于整个区间 $a\leq t\leq b$ 上, 且满足初始条件 $x(t_0)=\eta$. 

现取 $\mathbf \varphi_0(t)=\eta$, 可构造皮卡逐步逼近向量函数序列:

$$
\begin{cases}
\mathbf \varphi_0(t)=\mathbf \eta &\\
\mathbf \varphi_k(t)=\mathbf \eta+\int_{t_0}^t\left[\mathbf A(s)\varphi_{k-1}(s)+\mathbf f(s)\right]{\mathrm d s}, &k=1,2,\dotsi\\
\end{cases}
$$


则称向量函数 $\mathbf \varphi_k(t)$ 为

$$
\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=\mathbf A(t)\mathbf x+\mathbf f(t)
$$

的第 $k$ 次近似解.

**例 4 ** 求方程组的初值问题

$$
\frac{\mathrm d \mathbf x}{\mathrm d t}=\mathbf x'=
\begin{bmatrix}
1&0\\
-1&1\\
\end{bmatrix}
x+
\begin{bmatrix}
0\\
1\\
\end{bmatrix}
$$

$$
\mathbf x{(0)}=
\begin{bmatrix}
0\\
0\\
\end{bmatrix}
$$

的二次近似解.

**解**
令

$$
\mathbf \varphi_0(t)=
\begin{bmatrix}
0\\
0\\
\end{bmatrix}
$$

$$
\mathbf \varphi_1(t)=
\begin{bmatrix}
0\\
0\\
\end{bmatrix}
+\int_{0}^{t}\left(
\begin{bmatrix}
1&0\\
-1&1\\
\end{bmatrix}
\begin{bmatrix}
0\\
0\\
\end{bmatrix}
+
\begin{bmatrix}
0\\
1\\
\end{bmatrix}
\right){\mathrm d t}
=\int_0^t\begin{bmatrix}
0\\
1\\
\end{bmatrix}{\mathrm d t}=
\begin{bmatrix}
0\\
t\\
\end{bmatrix}
$$

$$
\mathbf \varphi_2(t)=
\begin{bmatrix}
0\\
0\\
\end{bmatrix}
+\int_{0}^{t}\left(
\begin{bmatrix}

\end{bmatrix}

\end{bmatrix}
$$

**例$6$** 求解方程组

$$
\begin{cases}
\frac{\mathrm d x}{\mathrm d t}=y\\
\frac{\mathrm d y}{\mathrm d t}=\frac{y^2}{x}\\
\end{cases}
$$

**解:** 
代入第一式代入第二式:
$$
\begin{align}
\frac{\mathrm d y}{\mathrm d t}& =\frac{1}{x}y^2\\
\frac{\mathrm d y}{\mathrm d x}\frac{\mathrm d x}{\mathrm d t} & =\frac{1}{x}y^2
\end{align}
$$

可得:
$$ 
y =0, \text{ 或 }\frac{\mathrm d y}{\mathrm d x}-\frac{1}{x}y=0 
$$

由 $\frac{\mathrm d y}{\mathrm d x}-\frac{1}{x}y=0$, 可得
$$
y = c_1x
$$
进而可得 
$$
\begin{cases}
x=c_2e^{c_1t}\\
y=c_1c_2e^{c_1t}\\
\end{cases}
$$

另外由 $y =0$, 可得:

$$
\begin{cases}
x=c\\
y=0\\
\end{cases}
$$

所以方程组的解为
$$
\begin{cases}
x=c_2e^{c_1t}\\
y=c_1c_2e^{c_1t}\\
\end{cases}
$$
